

#### **Task 2 Technical Report**
* **Sampling Strategy:** To distill the refined dataset, a *Stratified Random Sampling* framework was leveraged to target exactly 12,000 narratives. A simple random sample would introduce severe category risk due to the underlying class imbalances established in our EDA. Stratification ensures that minor classes (like Personal Loans) maintain their proportional weight (~6% of the vector index space), preventing the RAG system from developing a retrieval bias towards high-volume classes like Credit Cards.

* **Chunking Approach:** The pipeline utilizes LangChain's `RecursiveCharacterTextSplitter` configured to a strict window of `512` characters with a `50`-character recursive overlap. This specific size splits long, right-skewed narrative complaints into semantic windows containing approximately 2–3 sentences. This prevents the "dilution" of fine-grained details during vector pooling. The 50-character buffer handles overlapping boundaries, ensuring that critical data points (such as dates, account names, or specific violation keywords) split across boundaries are not lost to retrieval. The pipeline processed 12,000 sampled rows into exactly 34,605 clean text nodes.

* **Embedding Model Choice:** The open-source `all-MiniLM-L6-v2` transformer model was selected. This model compresses text chunks into compact 384-dimensional vectors. It was chosen because it represents an outstanding balance between local computational throughput and semantic accuracy. It requires minimal hardware memory footprint while maintaining high score metrics for capturing the natural, messy vocabulary typical of consumer financial disputes.

* **Indexing Architecture:** Vector search is managed using native `FAISS` running an Inner Product space (`IndexFlatIP`). Because all generated vectors are pre-normalized, calculating the inner product functions as a high-speed Cosine Similarity retrieval mechanism. Traceability is guaranteed by serializing an isolated mapping index (`metadata.pkl`), which anchors every generated chunk directly to its parent `Complaint ID` and `Product_Category` for accurate source-attribution during generation.
